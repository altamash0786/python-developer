```python
import tkinter as tk
from tkinter import messagebox

class Student:
    def __init__(self, name, roll, course, age):
        self.name = name
        self.roll = roll
        self.course = course
        self.age = age

class StudentManagementSystem:
    def __init__(self, root):
        self.students = {}
        self.root = root
        self.root.title("Student Management System")


        self.name_label = tk.Label(root, text="Name")
        self.name_label.grid(row=0, column=0)
        self.name_entry = tk.Entry(root)
        self.name_entry.grid(row=0, column=1)

        self.roll_label = tk.Label(root, text="Roll Number")
        self.roll_label.grid(row=1, column=0)
        self.roll_entry = tk.Entry(root)
        self.roll_entry.grid(row=1, column=1)

        self.course_label = tk.Label(root, text="Course")
        self.course_label.grid(row=2, column=0)
        self.course_entry = tk.Entry(root)
        self.course_entry.grid(row=2, column=1)

        self.age_label = tk.Label(root, text="Age")
        self.age_label.grid(row=3, column=0)
        self.age_entry = tk.Entry(root)
        self.age_entry.grid(row=3, column=1)


        self.add_button = tk.Button(root, text="Add Student", command=self.add_student)
        self.add_button.grid(row=4, column=0, pady=10)

        self.display_button = tk.Button(root, text="Display Students", command=self.display_students)
        self.display_button.grid(row=4, column=1, pady=10)

        self.search_button = tk.Button(root, text="Search Student", command=self.search_student)
        self.search_button.grid(row=5, column=0, pady=10)

        self.delete_button = tk.Button(root, text="Delete Student", command=self.delete_student)
        self.delete_button.grid(row=5, column=1, pady=10)

        self.update_button = tk.Button(root, text="Update Student", command=self.update_student)
        self.update_button.grid(row=6, column=0, pady=10)

        self.exit_button = tk.Button(root, text="Exit", command=root.quit)
        self.exit_button.grid(row=6, column=1, pady=10)

    def add_student(self):
        name = self.name_entry.get()
        roll = self.roll_entry.get()
        course = self.course_entry.get()
        age = self.age_entry.get()

        if roll in self.students:
            messagebox.showwarning("Error", "Student with this Roll Number already exists!")
        else:
            self.students[roll] = Student(name, roll, course, age)
            messagebox.showinfo("Success", "Student added successfully!")

    def display_students(self):
        if not self.students:
            messagebox.showinfo("Info", "No students to display.")
        else:
            display_window = tk.Toplevel(self.root)
            display_window.title("Student Details")
            for idx, student in enumerate(self.students.values(), start=1):
                details = f"{idx}. Name: {student.name}, Roll: {student.roll}, Course: {student.course}, Age: {student.age}"
                tk.Label(display_window, text=details).pack()

    def search_student(self):
        roll = self.roll_entry.get()
        if roll in self.students:
            student = self.students[roll]
            messagebox.showinfo("Student Found", f"Name: {student.name}, Roll: {student.roll}, Course: {student.course}, Age: {student.age}")
        else:
            messagebox.showerror("Error", "Student not found.")

    def delete_student(self):
        roll = self.roll_entry.get()
        if roll in self.students:
            del self.students[roll]
            messagebox.showinfo("Success", "Student deleted successfully!")
        else:
            messagebox.showerror("Error", "Student not found.")

    def update_student(self):
        roll = self.roll_entry.get()
        if roll in self.students:
            name = self.name_entry.get()
            course = self.course_entry.get()
            age = self.age_entry.get()

            self.students[roll].name = name
            self.students[roll].course = course
            self.students[roll].age = age

            messagebox.showinfo("Success", "Student details updated!")
        else:
            messagebox.showerror("Error", "Student not found.")



if __name__ == "__main__":
    root = tk.Tk()
    app = StudentManagementSystem(root)
    root.mainloop()
```
